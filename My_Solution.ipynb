2. Write a function that filters the dataset for questions that contains all of the words in a list of words. For example, when the list `["King", "England"]` was passed to our function, the function returned a DataFrame of 49 rows. Every row had the strings `"King"` and `"England"` somewhere in its `" Question"`.

   Test your function by printing out the column containing the question of each row of the dataset.

````python
def find(words):
    mask = jeopardy.Question.apply(lambda row : all(word in row for word in words)) #Using a boolean mask to label rows with True/False based on if they meet the criteria stated
    return jeopardy[mask] #Returns the True values from our Data Frame
````

3. Test your original function with a few different sets of words to try to find some ways your function breaks. Edit your function so it is more robust.

   For example, think about capitalization. We probably want to find questions that contain the word `"King"` or `"king"`.
   
   You may also want to check to make sure you don't find rows that contain substrings of your given words. For example, our function found a question that didn't contain the word `"king"`, however it did contain the word `"viking"` &mdash; it found the `"king"` inside `"viking"`. Note that this also comes with some drawbacks &mdash; you would no longer find questions that contained words like `"England's"`.

````python
def filter_by_words(words):
    normalized_words = [word.lower() for word in words] #Converts all the words we are looking for to lowercase for comparison
    
    def check_question(question):
        question_words = question.lower().split() #Converts all the words that we are comparing against to lowercase and splits them up into a list so we compare against the entire word
        return all(word in question_words for word in normalized_words) #Retruns True if the word in question_words is a word in normalised_words
    
    return jeopardy[jeopardy["Question"].apply(check_question)] #Applies our function to the Question column specifically and returns all the True rows

#To improve this function and make it more robust we can use string.punctuation to remove punctuation from our text before splitting
````

````python
import random
import string
````

````python
def lower_strip(words):
    words = words.lower()
    words = words.strip() #Removes any whitespace
    return words

def clean(text):
    cleaned_text = ''
    text = lower_strip(text)
    for char in text:
        if char not in string.punctuation: #Removes any punctuation so that we can accurately compare question and answer
            cleaned_text += char
    return cleaned_text

def quiz():
    random_index = random.randint(0, len(jeopardy)-1) #Picks a random row for our quiz
    random_question = jeopardy.iloc[random_index]['Question'] #Selects and stores the Question from that row
    print(random_question)
    answer = input() #Stores the users input for comparison
    true_answer = jeopardy.iloc[random_index]['Answer'] #Stores the answer from our Data Frame 
    #Using our previous functions to clean our answers for a more accurate comparison
    answer = clean(answer)
    answer_for_comparison = clean(true_answer)
    #Uses an if statement to compare answers and output whether the user is correct or incorrect
    if answer == answer_for_comparison:
        print('Correct!')
    else:
        print('That is the wrong answer')
        print('The correct answer is: ' + true_answer)
````
